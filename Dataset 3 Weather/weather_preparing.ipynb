```python
import numpy as np
import pandas as pd
import glob
```

# Loading Data

```python
path = r'C:\Users\alsah\Documents\Module03\Project - Scraping\WeatherDatasets' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df
```

# Feature Selection

```python
df_test = df[['date', 'Temp', 'Condition', 'Wind Speed']].copy()
```

```python
df_test = df_test.groupby('date').agg({'Temp': ['min', 'max'], 'Condition': ['sum']}).reset_index()
```

```python
df_test = pd.DataFrame({'date':df_test['date'],
                        'temp_min':df_test['Temp']['min'],
                        'temp_max':df_test['Temp']['max'],
                        'is_rain':df_test['Condition']['sum'].apply(lambda x: 'Rain' in x)})
```

```python
df_test
```

```python
df_test.to_csv('weather cleaned 2018-1-1 to 2021-9-30.csv')
```
